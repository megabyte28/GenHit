```python
from ultralytics import YOLO
import os

# 1. Load the Nano model (fastest for hackathons)
model = YOLO('yolov8n.pt')

# 2. Train the model
# We point directly to the data.yaml file inside your downloaded folder
model.train(
    data=os.path.join(dataset.location, "data.yaml"),
    epochs=25,    # It will read the data 25 times
    imgsz=640,    # Standard image size
    plots=True    # Generates charts so you can see the progress
)
```

```python
from google.colab import files
import os

# Updated path based on your logs (train4)
model_path = '/content/runs/detect/train4/weights/best.pt'

if os.path.exists(model_path):
    print("Model mil gaya! Downloading now...")
    files.download(model_path)
else:
    # Agar ye fir bhi na chale, toh ye code saare folders check karega
    print("Manual path fail hua, scanning for latest 'best.pt'...")
    import glob
    all_weights = glob.glob('/content/runs/detect/train*/weights/best.pt')
    if all_weights:
        latest_model = max(all_weights, key=os.path.getmtime)
        print(f"Latest model found at: {latest_model}")
        files.download(latest_model)
    else:
        print("Bhai, file sach mein nahi mil rahi. Check kar sidebar mein 'runs' folder hai ya nahi.")
```
